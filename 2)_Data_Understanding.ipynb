## **BALANCED vs IMBALANCED DATA**



---

* The data might be a CATEGORICAL OR CONTINOUS in a datasets that we get.

* While it is a Categorical Data when the datasets has 2 classes, if neither one is less than 30% of accomodate,i.e: **both the classes has an equal accomodation** then it is said to be **IMBALANCED DATA.**

* While it is a Categorical Data when the datasets has 2 classes, if either one is **less than 30%** of accomodate, then it is said to be **IMBALANCED DATA.**

* So, if a either category has less than 30% accomodation it is said to be a **THUMB RULE**. Globally by many years experienced data scientists have named it as a **THUMB RULE**. Though it is not a proven statistics percentage.

* While it is a Continuos dataset, we can obtain a GRAPH of distribution. If we get a NORMAL DISTRIBUTION, i.e: summetrical distributions of data, it is said to be a Balanced dataset. When it is not, then we can colclude that the datasets might be a IMBALANCED OR Rare event.

## **CROSS SECTIONAL vs TIME SERIES vs PANEL / LONGITUDINAL DATA**

---

* **CROSS SECTIONAL DATA** where
The DATE AND TIME is unimportant and

Sequence of order is unimportant.

In Cross sectional data more than one column/variable/feature will be there.

:NC = NOT COVID ; C = COVID

* **TIME SERIES DATA** where the
The DATE AND TIME is Extremely important and
